```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def plot_distribution(distribution):
    px = pd.Series(distribution)
    px = px.value_counts().sort_index() / len(px)
    plt.plot(px.index, px, marker='o')
    plt.xlabel('k')
    plt.ylabel('P(X=k)')
size = 1000
```

## 二項分布
結果が2通りの試行(ベルヌーイ試行)を$n$回行った時に$k$回成功する確率を示す確率分布．  
$$P(X=k)={}_n C_{k} p^k (1-p)^{n-k}$$
期待値は$np$，分散は$np(1-p)$となる．

```python
# arguments: (n, p, size) where p is probability of success
plot_distribution(np.random.binomial(10, .2, size))
plot_distribution(np.random.binomial(10, .5, size))

plt.show()
```

## ポアソン分布  
標本となる期間で平均$\lambda$回起こる事象が$k$回起きる確率を示す確率分布．
$$P(X=k) = \frac{e^{\lambda}\lambda^k}{k!}$$  
期待値は$\lambda$，分散は$\lambda$となる．

```python
plot_distribution(np.random.poisson(2,size))
plot_distribution(np.random.poisson(4,size))

plt.show()
```

## 幾何分布
成功確率$p$の独立なベルヌーイ試行を繰り返し，初めて成功するまでの試行回数$X$が従う確率分布．
$$P(X=k) = (1-p)^{k-1} p$$
期待値は$\frac{1}{p}$，分散は$\frac{1-p}{p^2}$となる．

```python
plot_distribution(np.random.geometric(1/6, size))
plot_distribution(np.random.geometric(1/4, size))

plt.show()
```
