```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def plot_distribution(distribution, color=None):
    px = pd.Series(distribution)
    px = pd.cut(px, 100)
    px = px.value_counts().sort_index() / len(px)
    px.index = px.index.map(lambda x: x.left).values.astype(np.float32)
    plt.plot(px.index, px.values, c=color)
    plt.grid()
    plt.xlabel('x')
    plt.ylabel('P(X=x)')
size = int(1e5)
```

```python
for deg in [1,2,3,5,10]:
    plot_distribution(np.random.chisquare(deg, size))
    
plt.title('χ-squared distribution')
plt.xlim(0, 20)
plt.ylim(0, 0.10)
plt.legend(['k=1','2','3','5','10'])
plt.show()
```
